M√≥dszerek kigy≈±jt√©se

Tev√©kenys√©gek oszlopban milyen ig√©k √©s f≈ënevek jelennek meg gyakran.

```python
El≈ësz√∂r a legkisebb (89) adatb√°zison pr√≥b√°ltam ki az √∂tletemet.
```

```python
import pandas as pd
import spacy
from collections import Counter

nlp = spacy.load("hu_core_news_lg")

# Excel beolvas√°sa
df = pd.read_excel("CERV-F3-2024_89db_text_lemmatizalt.xlsx")

# Csak a "Tev√©kenys√©gek" oszlopot vessz√ºk
szovegek = df["Tev√©kenys√©gek"].dropna().astype(str)

# Tokeniz√°l√°s √©s POS-sz≈±r√©s
igek = []
fonevek = []

for szoveg in szovegek:
    doc = nlp(szoveg)
    for token in doc:
        if token.pos_ == "VERB" and not token.is_stop:
            igek.append(token.lemma_.lower())
        elif token.pos_ == "NOUN" and not token.is_stop:
            fonevek.append(token.lemma_.lower())

# Gyakoris√°gi list√°k
print("üîπ Leggyakoribb ig√©k:")
print(Counter(igek).most_common(30))

print("\nüîπ Leggyakoribb f≈ënevek:")
print(Counter(fonevek).most_common(30))
```

Az leggyakoribb 30 ige √©s f≈ën√©v √°tn√©z√©se ut√°n arra jutottam, hogy a f≈ënevek √°tn√©z√©se c√©lravezet≈ëbb lehet. A m√≥dszerekhez ink√°bb a f≈ëneveken kereszt√ºl jutunk el, mint az ig√©ken kereszt√ºl.

Kiirattam azokat a f≈ëneveket amik legal√°bb k√©tszer fordulnak el≈ë:

```python
fonev_gyakorisag = Counter(fonevek)

# csak azok, amik legal√°bb k√©tszer fordulnak el≈ë
fonevek_legalabb_ketszer = {szo: db for szo, db in fonev_gyakorisag.items() if db >= 2}

# eredm√©nyek rendez√©se √©s megjelen√≠t√©se
fonevek_legalabb_ketszer = dict(sorted(fonevek_legalabb_ketszer.items(), key=lambda x: x[1], reverse=True))

print(f"{len(fonevek_legalabb_ketszer)} f≈ën√©v fordul el≈ë legal√°bb k√©tszer:\n")
for szo, db in fonevek_legalabb_ketszer.items():
    print(f"{szo}: {db}")
```
